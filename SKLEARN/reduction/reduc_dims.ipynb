```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
```

```python
digits = load_digits()
images = digits.images
X = digits.data
y = digits.target

print(X.shape)
```

```python
model = PCA(n_components=2)

X_reduced = model.fit_transform(X)

print(X_reduced.shape)
```

```python
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='jet')
plt.colorbar()
```

# Using it for reducing time

```python
model = PCA(n_components=64)

X_reduced = model.fit_transform(X)
```

```python
np.cumsum(model.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(model.explained_variance_ratio_))
```

```python
np.argmax(np.cumsum(model.explained_variance_ratio_) > 0.95)
```

```python
model = PCA(n_components=.2)
```

```python
X_compress = model.fit_transform(X)
X_decompress = model.inverse_transform(X_compress)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(X[0, :].reshape(8, 8), cmap='gray')
plt.title('Original')
plt.subplot(1, 2, 2)
plt.imshow(X_decompress[0, :].reshape(8, 8), cmap='gray')
plt.title('Compressed')
```
