```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=100, centers=3, cluster_std=0.4, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=30)
```

```python
model = KMeans(n_clusters=3, init='random', n_init=10, max_iter=10, tol=1e-04, random_state=0)
model.fit(X)
y_pred = model.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=30)
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=250, marker='*', c='red')

print('Distortion: %.2f' % model.inertia_)
print('SSE: %.2f' % model.inertia_)
print('Centers: \n', model.cluster_centers_)
print('Labels: \n', model.labels_)
print('Predict: \n', model.predict(X))
print('Score: %.2f' % model.score(X))
# Test new data
new_data = np.array([[-2, 3], [2, 2]])
print('New data: \n', model.predict(new_data))

plt.plot(new_data[:, 0], new_data[:, 1], 'b*', markersize=20)
plt.show()
```

```python
inertia = []
for k in range(1, 20):
    model = KMeans(n_clusters=k, init='random', n_init=10, max_iter=10, tol=1e-04, random_state=0)
    model.fit(X)
    inertia.append(model.inertia_)
```

```python
plt.plot(range(1, 20), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```
